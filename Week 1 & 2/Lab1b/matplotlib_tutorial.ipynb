### Control the plot size

Specify the size with plt.figure() before creating the plot.

```python
plt.figure(figsize=(4,4))
# make a line plot: generate x 50 values evenly spaced between +1 and +50
x = np.linspace(1, 50)
# Evaluate the log
y = np.log(x)
# plot it!
plt.plot(x, y)
plt.show()
```

### Plot multiple log curves in same plot

```python
# We can plot a few more on the same plot
plt.figure(figsize=(4,4))
# make a line plot: generate x 50 values evenly spaced between +1 and +50
x = np.linspace(1, 50)
# Evaluate the log
y1 = np.log(x)
y2 = np.log(10*x)
y3 = np.log(50*x)

# plot it!
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

### Customize colors

```python
# We can plot a few more on the same plot
plt.figure(figsize=(4,4))
# make a line plot: generate x 50 values evenly spaced between +1 and +50
x = np.linspace(1, 50)
# Evaluate the log
y1 = np.log(x)
y2 = np.log(10*x)
y3 = np.log(50*x)

# plot it!
plt.plot(x, y1, color='r')
plt.plot(x, y2, color='g')
plt.plot(x, y3, color='k')  # black
plt.show()
```

### Use a qualitative (categorical) color palette

Options here: https://matplotlib.org/stable/tutorials/colors/colormaps.html#qualitative

Access list of RGB colors via: `plt.cm.PALETTENAME.colors`

```python
plt.cm.tab10.colors
```

```python
# We can plot a few more on the same plot
plt.figure(figsize=(4,4))
# make a line plot: generate x 50 values evenly spaced between +1 and +50
x = np.linspace(1, 50)
# Evaluate the log
y1 = np.log(x)
y2 = np.log(10*x)
y3 = np.log(50*x)

# plot it!
plt.plot(x, y1, color=plt.cm.tab10.colors[0])
plt.plot(x, y2, color=plt.cm.tab10.colors[1])
plt.plot(x, y3, color=plt.cm.tab10.colors[2])  # black
plt.show()
```

### Customize examples: line thickness (aka width) and style

```python
# We can plot a few more on the same plot
plt.figure(figsize=(4,4))
# make a line plot: generate x 50 values evenly spaced between +1 and +50
x = np.linspace(1, 50)
# Evaluate the log
y1 = np.log(x)
y2 = np.log(10*x)
y3 = np.log(50*x)

# plot it!
plt.plot(x, y1, color=plt.cm.tab10.colors[0], linewidth=5, linestyle='dashed')
plt.plot(x, y2, color=plt.cm.tab10.colors[1], linewidth=2, linestyle='dotted')
plt.plot(x, y3, color=plt.cm.tab10.colors[2], linewidth=0.5)  # black
plt.show()
```

### Set title, axis viewing ranges, legend

```python
# We can plot a few more on the same plot
plt.figure(figsize=(4,4))
# make a line plot: generate x 50 values evenly spaced between +1 and +50
x = np.linspace(1, 50)
# Evaluate the log
y1 = np.log(x)
y2 = np.log(10*x)
y3 = np.log(50*x)

# plot it!
plt.plot(x, y1, color=plt.cm.tab10.colors[0], linewidth=5, linestyle='dashed', label='log(x)')
plt.plot(x, y2, color=plt.cm.tab10.colors[1], linewidth=2, linestyle='dotted', label='log(10x)')
plt.plot(x, y3, color=plt.cm.tab10.colors[2], linewidth=0.5, label='log(50x)')  # black
plt.title('My amazing log curves!')
plt.xlabel('x values')
plt.ylabel('log(x)')
plt.xlim(-10, 60)
plt.ylim(-5, 10)
plt.legend()
plt.show()
```

## Subplots

**Matplotlib is powerful!** 

- Let's create a grid of stacked plots with `plt.subplots`.
- Let's make it 2 rows and 1 col. 
- Let's put the log plot on top and a scatter plot of random data below

```python
# The "grid" object of 2 subplots is called a Figure object
# The subplots are called Axes
# Each axis acts like plt above, except the plot decoration commands
# usually have a set_ before them. e.g. ax.set_xlabel('My x label')
# Control figure size with figsize optional parametr of plt.subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(5, 10))
# set the title
# make a line plot: generate x 50 values evenly spaced between +1 and +50
x = np.linspace(1, 50)
# Evaluate the log
y1 = np.log(x)
y2 = np.log(10*x)
y3 = np.log(50*x)

# plot it!
axes[0].plot(x, y1, color=plt.cm.tab10.colors[0], linewidth=5, linestyle='dashed', label='log(x)')
axes[0].plot(x, y2, color=plt.cm.tab10.colors[1], linewidth=2, linestyle='dotted', label='log(10x)')
axes[0].plot(x, y3, color=plt.cm.tab10.colors[2], linewidth=0.5, label='log(50x)')  # black
axes[0].set_title('My amazing log curves!')
axes[0].set_xlabel('x values')
axes[0].set_ylabel('log(x)')
axes[0].set_xlim(-10, 60)
axes[0].set_ylim(-5, 10)
# axes[0].legend()
# plt.show()

# Generate random data in both x and y
x_r = np.random.random([50,])
y_r = np.random.random([50,])

# Make scatter plot
axes[1].scatter(x_r, y_r)
axes[1].set_title('Scatter plot!')
plt.show()
```

```python
# The "grid" object of 2 subplots is called a Figure object
# The subplots are called Axes
# Each axis acts like plt above, except the plot decoration commands
# usually have a set_ before them. e.g. ax.set_xlabel('My x label')
# Control figure size with figsize optional parametr of plt.subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(5, 10))
# set the title
# make a line plot: generate x 50 values evenly spaced between +1 and +50
x = np.linspace(1, 50)
# Evaluate the log
y1 = np.log(x)
y2 = np.log(10*x)
y3 = np.log(50*x)

# plot it!
axes[0,0].plot(x, y1, color=plt.cm.tab10.colors[0], linewidth=5, linestyle='dashed', label='log(x)')
axes[0,0].plot(x, y2, color=plt.cm.tab10.colors[1], linewidth=2, linestyle='dotted', label='log(10x)')
axes[0,0].plot(x, y3, color=plt.cm.tab10.colors[2], linewidth=0.5, label='log(50x)')  # black
axes[0,0].set_title('My amazing log curves!')
axes[0,0].set_xlabel('x values')
axes[0,0].set_ylabel('log(x)')
axes[0,0].set_xlim(-10, 60)
axes[0,0].set_ylim(-5, 10)
axes[0,0].legend()
# plt.show()

# Generate random data in both x and y
x_r = np.random.random([50,])
y_r = np.random.random([50,])

# Make scatter plot
axes[0,1].scatter(x_r, y_r)
axes[0,1].set_title('Scatter plot!')
plt.show()
```
